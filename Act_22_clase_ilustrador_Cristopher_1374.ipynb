```python
class Ilustrador:
    def __init__(self, nombre, estilo):
        self.nombre = nombre
        self.estilo = estilo
        self.dibujando = False

    def empezar_dibujar(self):
        self.dibujando = True
        print(f"El ilustrador {self.nombre} está empezando a dibujar en estilo {self.estilo}.")

# Uso de la clase
ilustrador1 = Ilustrador("Cristopher", "Digital")
print(f"Ilustrador: {ilustrador1.nombre}, Estilo: {ilustrador1.estilo}")
ilustrador1.empezar_dibujar()

ilustrador2 = Ilustrador("Maria", "Tradicional")
print(f"Ilustrador: {ilustrador2.nombre}, Estilo: {ilustrador2.estilo}")
ilustrador2.empezar_dibujar()

print(" creador de software cristopher Lopez NL = 33")
```
